```python
# Importing the json file with all 50 plans
import pandas as pd
dataset=pd.read_json("final.json",lines=True)
df= pd.DataFrame(dataset)
print(df)
```

```python
# Regular Expressions(re)
import re
#Natural Language toolkit(nltk)
import nltk
# Stopword are all articles used in sentences('the','is','was' etc which doesnt help us )
nltk.download('stopwords')
from nltk.corpus import stopwords

corpus=[]
#There are 50 plans in the dataset
for i in range(0,50):
    #Replacing everthing except a-z and A-z with space(' ')
    plan=re.sub('[^a-zA-Z1-9]'," ",df['document'][i])
    plan=plan.lower()
    #spliting each word
    plan = plan.split()
    all_stopwords=stopwords.words('english')
    
    plan=[word.replace(word,word) for word in plan if not word in set(all_stopwords)]
    #Joining all the split words in the plans with a ' ' between them
    plan=' '.join(plan)
    #Appending each plan to corpus
    corpus.append(plan)
```

```python
# Printing the first plan alone
# Editing the corpus index here will give us the different table and matrix representation of the different plans
text=corpus[0]
print(text)
```

```python
import string 
import spacy 
import numpy as np 
import math 
from tqdm import tqdm 

from spacy.matcher import Matcher 
from spacy.tokens import Span 
from spacy import displacy

nlp = spacy.load("en_core_web_sm")



# creating  a spaCy object 
doc = nlp(text)
nlp = spacy.load("en_core_web_sm")
           
```

```python
from spacy.lang.en import English
from spacy.pipeline import EntityRuler

# Using the spacy entity ruler to name the differrnt rooms in the plan with entity type -rooms
nlp = English()
ruler = EntityRuler(nlp)
patterns = [{"label": "rooms", "pattern": "livingroom1"},{"label": "rooms", "pattern": "bedroom1"},{"label": "rooms", "pattern": "bedroom2"},{"label": "rooms", "pattern": "bedroom3"},{"label": "rooms", "pattern": "bedroom4"},
            {"label": "rooms", "pattern": "washroom1"},{"label": "rooms", "pattern": "washroom2"},{"label": "rooms", "pattern": "washroom3"},{"label": "rooms", "pattern": "kitchen1"},
            {"label": "rooms", "pattern": "balcony1"},{"label": "rooms", "pattern": "balcony2"},{"label": "rooms", "pattern": "balcony3"},{"label": "rooms", "pattern": "study"}]
ruler.add_patterns(patterns)
nlp.add_pipe(ruler)

doc = nlp(text)
print([(ent.text, ent.label_) for ent in doc.ents])

#Printing all the words along with their entity names
for tok in doc:
    print(tok.text,tok.ent_type_)
```

```python


#defining the different patterns in the plans to match and extract the information we need
pattern1 = [{'ENT_TYPE':'rooms'}, 
           {'LOWER': 'adjacent'}, 
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'}
           ]
pattern2= [{'ENT_TYPE':'rooms'}, 
           {'LOWER': 'adjacent'}, 
           {'ENT_TYPE':'rooms'}
           ]
pattern3= [{'ENT_TYPE':'rooms'}, 
           {'LOWER': 'next'}, 
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'}
           ]
pattern4= [{'ENT_TYPE':'rooms'}, 
           {'LOWER': 'next'}, 
           {'ENT_TYPE':'rooms'},
           ] 
pattern5=[{'ENT_TYPE':'rooms'},  
           {'ENT_TYPE':'rooms'},
           {'LOWER': 'connected'}
          ] 
pattern5=[{'ENT_TYPE':'rooms'},  
           {'ENT_TYPE':'rooms'},
          {'ENT_TYPE':'rooms'},  
           {'ENT_TYPE':'rooms'},
          {'ENT_TYPE':'rooms'},  
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},  
           {'ENT_TYPE':'rooms'},
           {'LOWER': 'connected'}
          ] 
pattern6 = [{'ENT_TYPE':'rooms'}, 
           {'LOWER': 'adjacent'}, 
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           ]
pattern7 = [{'ENT_TYPE':'rooms'}, 
           {'LOWER': 'adjacent'}, 
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'}
           ]
pattern8=[{'ENT_TYPE':'rooms'}, 
           {'LOWER': 'adjacent'}, 
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'}
           ]
pattern9=[{'ENT_TYPE':'rooms'},  
           {'ENT_TYPE':'rooms'},
          {'ENT_TYPE':'rooms'},  
           {'ENT_TYPE':'rooms'},
          {'ENT_TYPE':'rooms'},  
           {'ENT_TYPE':'rooms'},
           {'ENT_TYPE':'rooms'},  
           {'LOWER': 'connected'}
          ] 

# Adding the patterns to the Matcher class object 
matcher = Matcher(nlp.vocab) 
matcher.add("matching_1", None, pattern1) 
matcher.add("matching_2", None, pattern2) 
matcher.add("matching_3", None, pattern3) 
matcher.add("matching_4", None, pattern4)
matcher.add("matching_5", None, pattern5)
matcher.add("matching_6", None, pattern6)
matcher.add("matching_7", None, pattern7)
matcher.add("matching_8", None, pattern8)
matches = matcher(doc) 

print(len(matches))

span=[]
for i in range(0,len(matches)):
    span.append(doc[matches[i][1]:matches[i][2]])
print(span)

```

```python
# Identifying the different unique rooms in the mentioned plan
allrooms=[]
for ent in doc.ents:
        allrooms.append((ent.text))
allrooms=np.array(allrooms)
uniquerooms=np.unique(allrooms)
uniquerooms=uniquerooms.reshape(len(uniquerooms),1)
print(uniquerooms)
```

```python
# Function to check if a word as entity type and return 1 if it does and 0 if it doesnt
#Using this function to remove all the connector words like 'connected','next to','adjacent' from span to get the desired table format
def check(word):
        if  word.ent_type_:
            return(word)
```

```python
links=[]
for i in range (0,len(span)):
    for j in range (0,len(span[i])-1):
        if check(span[i][j]) and check(span[i][j+1]):
                links.append([(span[i][j]),(span[i][j+1])])
        elif check(span[i][j]) and not check(span[i][j+1]):
                links.append([(span[i][j]),(span[i][j+2])])
         
```

```python
links1=np.array(links)
links=pd.DataFrame(links1,columns=['Rooms','link'])
display(links)
```

```python
#Creating a array with len of uniqueroom rows and len of uniquerooms columns
no_of_rooms=len(uniquerooms)
print(no_of_rooms)
matrix=np.zeros((no_of_rooms,no_of_rooms))
```

```python
#Creating a pandas dataframe of the matrix with all the names of the unique rooms being the names and the columns of the matrix
matrix=pd.DataFrame(matrix,uniquerooms,uniquerooms)
print(matrix)
```

```python
#Updating the values in the matrix
for rooms, link in links1:
         matrix[(str(rooms),)][(str(link),)]=1
         matrix[(str(link),)][(str(rooms),)]=1
# Displaying the matrix Representation
display(matrix) 
```
